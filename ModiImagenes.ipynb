```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('misteranderson.png')

R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
xx = plt.imshow(imgGray, cmap='gray')
plt.show()
```

### RGB

```python
mister_rgb=io.imread('misteranderson.png')/255.0 # imread lee las imagenes con los pixeles codificados como enteros 
# en el rango 0-255. Por eso la convertimos a flotante y en el rango 0-1
plt.imshow(mister_rgb) # no es necesario el vmin/vmax para imgenes a color

print("Dimensiones de la imagen:")
print(mister_rgb.shape)
```

```python
plt.imshow(mister_rgb[:,:,0],vmin=0,vmax=1)
plt.title("Canal Rojo")
plt.figure()
plt.imshow(mister_rgb[:,:,1],vmin=0,vmax=1)
plt.title("Canal Verde")
plt.figure()
plt.imshow(mister_rgb[:,:,2],vmin=0,vmax=1)
plt.title("Canal Azul")
```

```python
mister_red=np.copy(mister_rgb) # creo una copia de la imagen para preservar la original
mister_red[:,:,1]=0
mister_red[:,:,2]=0
plt.title("Mister_Anderson canal rojo")
plt.imshow(mister_red)
```

```python
mister_red_green=np.copy(mister_rgb) # creo una copia de la imagen para preservar la original
mister_red_green[:,:,2]=0
plt.title("Mister_Anderson sin canal azul")
plt.imshow(mister_red_green)
```
